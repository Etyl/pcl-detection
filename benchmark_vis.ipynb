```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
augmentation_benchmark_large = {
    "No augmentation": (0.5997,0.0165),
    "GPT": (0.6021,0.0211),
    "Synonym + CWE": (0.5638,0.0118),
}

augmentation_benchmark_base = {
    "No augmentation": (0.5904,0.0081),
    "GPT": (0.578,0.0072),
    "Synonym + CWE": (0.5805,0.0109),
}

wrs_benchmark_large = {
    "No WRS": (0.5995539882,0.02152478181),
    "sqrt WRS": (0.6021,0.0211),
    "linear WRS": (0.5890,0.0037),
}
```

```python
def plot_benchmark(benchmark):
    labels = list(benchmark.keys())
    means = [v[0] for v in benchmark.values()]
    stds = [v[1] for v in benchmark.values()]

    plt.figure(figsize=(8, 6))

    # Set positions for bars manually to ensure even spacing
    x_positions = range(len(labels))

    # Plot mean points with whiskers
    plt.errorbar(x_positions, means, yerr=stds, fmt='o', color='black', capsize=10, 
                label='Mean Â± Std', markersize=10)

    # Set the x-tick positions and labels
    plt.xticks(x_positions, labels)

    plt.ylabel("F1 Score")
    plt.ylim(0.5, 0.65)
    plt.grid(axis='y', linestyle='--', alpha=0.6)  # Only horizontal grid lines
    plt.legend()

    # Add some padding to x-axis
    plt.xlim(-0.5, len(labels)-0.5)

    plt.tight_layout()  # Adjust layout to ensure everything fits
    plt.show()
```

```python
plot_benchmark(augmentation_benchmark_large)
```

```python
plot_benchmark(augmentation_benchmark_base)
```

```python
plot_benchmark(wrs_benchmark_large)
```
